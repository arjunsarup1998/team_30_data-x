```python
import numpy as np 
import pandas as pd
import json
import time
from xgboost import XGBClassifier
```

```python
image_DF = pd.read_csv('/Users/aceman/Desktop/alia_image_output.csv', names=['name', 'objects'])
image_DF
```

```python
with open('./aliaabhatt.json') as json_file:  
    AB_data = json.load(json_file)
```

```python
data = [['DOW', 'day', 'month', 'year', 'time', 'timezone', 'epoch', 'num_likes', 'num_comments', 'url']]
```

```python
for entry in AB_data['GraphImages']: 
    if entry['is_video'] == False:
        epoch = entry['taken_at_timestamp']
        t = time.strftime("%a, %d %b %Y %H:%M:%S %Z", time.localtime(epoch))
        row = t.split(' ')
        row[0] = row[0][:len(row[0])-1]
        row.append(epoch)
        row.append(entry['edge_media_preview_like']['count'])
        row.append(entry['edge_media_to_comment']['count'])
        temp_url = entry['display_url'].split('.jpg')[0].split('/')[8] + '.jpg'
        row.append(temp_url)
        data.append(row)
```

```python
AB_df = pd.DataFrame(data, columns=['DOW', 'day', 'month', 'year', 'time', 'timezone', 'epoch', 'num_likes', 'num_comments', 'url'])
AB_df = AB_df.drop([0])
```

```python
AB_df.head()
```

```python
new_AB_df = AB_df.merge(image_DF, how='inner', left_on='url', right_on='name')
new_AB_df = new_AB_df.drop(columns=['url'])
num_persons = []
num_cats = []
num_dogs = []
for i in new_AB_df['objects']:
    count = i.count('person')
    if count > 0:
        num_persons.append(1)
    else:
        num_persons.append(0)
    
new_AB_df['is_person'] = num_persons
```

```python
curr_index = 1
previous_likes = np.array([])
while (curr_index <= len(new_AB_df)):
    end_index = min(len(new_AB_df) + 1, curr_index + 10)
    previous_likes = np.append(previous_likes, new_AB_df['num_likes'][curr_index:end_index].mean())
    curr_index=curr_index + 1
    
previous_likes[len(new_AB_df) - 1] = 0
new_AB_df['previous_likes'] = previous_likes
```

```python
from sklearn.linear_model import LinearRegression
```

```python
avg_likes = new_AB_df['num_likes'].mean()
```

```python
new_AB_df['datetime'] = pd.to_datetime(new_AB_df['time'])
new_AB_df['hour'] = [i.hour for i in new_AB_df['datetime']]
```

```python
new_AB_df['avg_likes'] = pd.Series(np.ones(new_AB_df.shape[0])*avg_likes)
new_AB_df['follower_count'] = pd.Series(np.ones(new_AB_df.shape[0])*30968015)
days = pd.get_dummies(new_AB_df['DOW'])
```

```python
months = pd.get_dummies(new_AB_df['month'])
```

```python
new_AB_df = new_AB_df.merge(days, left_index = True, right_index = True)
```

```python
new_AB_df = new_AB_df.merge(months, left_index = True, right_index = True)
```

```python
new_AB_df = new_AB_df.dropna()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
X = new_AB_df.loc[:, ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun', 'epoch', 'previous_likes', 'is_person', 'hour']]
Y = new_AB_df.loc[:, 'num_likes']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, shuffle = True, random_state=100)
LinearRegressionModel = linear_model.LinearRegression() 
LinearRegressionModel.fit(x_train, y_train)
print("Training: ", LinearRegressionModel.score(x_train, y_train))
print("Testing: ", LinearRegressionModel.score(x_test, y_test))
```
